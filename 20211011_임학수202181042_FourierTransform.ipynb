### Fourier Transform 4

#### Example

$ x_1 (t) = r(t) - 2r ( t - 0.5) + r(t-1), \quad \quad \Omega_0 = 2 \pi$

$ = \sum X_k {e}^{jk \Omega_0 k} $ <br><br>
$ \therefore X_k = \frac {1}{T} \int_{-\frac{T}{2}}^{0} -2t e^{jk \Omega_0 t} dt + \frac {1}{T} \int_{0}^{\frac{T}{2}} 2t e^{-jk \Omega_0 t} dt $ <br><br>
$ = \int_{0}^{\frac{1}{2}} 2t e^{jk\Omega_0 t}dt + \int_{0}^{\frac{1}{2}} 2t e^{-jk\Omega_0 t}dt $<br><br>
$ = \int_{0}^{\frac{1}{2}} 4t \cos (k \Omega_0t)dt = \int_{0}^{\frac{1}{2}} 4t \cos (2\pi kt)dt$<br><br>
$ 4 \left[ \frac{t}{2 \pi k} \sin (2 \pi kt) + \frac{1}{4 \pi^2 k^2} \cos (2 \pi kt) \right]_{0}^{\frac{1}{2}} $ <br><br>
$ = 4 \frac{1}{4 \pi^2 {k}^{2}}(cos (\pi k -1)) $

$${X}_{k} = \frac{1}{\pi^2 k^2}((-1)^k-1)$$

${X}_{0} = \left( \int_{\frac{1}{2}}^{0}-t dt + \int_{\frac{1}{2}}^{0} t dt \right) = \left[ - \frac{t^2}{2} \right]_{-\frac{1}{2}}^{0} + \left[\frac{t^2}{2} \right]_{0}^{\frac{1}{2}} = \frac{1}{4}$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def r(t):
    return t * np.heaviside(t, 0)
```

```python
def X(k):
    return 1/2/np.pi**2/k**2 * ((-1)**k -1)
t = np.linspace(-10, 10, 1000)
plt.figure(figsize=(12,6))
plt.plot(t, r(t) - 2*r(t-0.5) + r(t-1))

plt.plot(t, 1/4*np.ones_like(t))
plt.plot(t, np.real(1/4*np.ones_like(t) + X(1) * np.exp(1j * 2*np.pi *t) + X(3) * np.exp(1j * 2 *np.pi *3 *t)))
```

```python
def X(k):
    return 1/2/np.pi**2/k**2 * ((-1)**k -1)

def x1(t):
    return r(t) - 2 * r(t-0.5) + r(t-1)
t = np.linspace(-3, 3, 1000)
plt.figure(figsize=(12,6))
plt.plot(t, x1(t) + x1(t-1) + x1(t-2))

plt.plot(t, 1/4*np.ones_like(t))

v = 1/4*np.ones_like(t)
for k in range (-5, 5):
    if k == 0:
        continue
    v = v + X(k) * np.exp(1j * 2* np.pi * k *t)

plt.plot(t, np.real(v))
```

```python
def X(k):
    return 2 * ((-1)**k) / np.pi / ( 1 - 4 * k**2 )

def x1(t):
    return 12 * np.abs(np.cos( 100 * np.pi * t))

t = np.linspace(-0.1, 0.1, 1000)
plt.figure(figsize = (12,6))
plt.plot(t, x1(t), label = '|cos|')

# plt.plot(t, 12 * 2 / np.pi * np.ones_like(t))

v = 12 * 2 / np.pi * np.ones_like(t)
for k in range(-14, 15):
    if k == 0:
        continue
    v = v + X(k) * 12 * np.exp(1j * 200 * np.pi * k * t) * 1 / np.sqrt( 1 + 0.01 * (200 * np.pi * k ) ** 2 )
plt.plot(t, np.real(v), label = "FS")
plt.legend()
```
