## Discrete Time Signals

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy import signal
import numpy as np
%matplotlib inline
from IPython.display import Image
```

### Example
$x(t) = 3 \cos \left( 2 \pi t + \frac{\pi}{4} \right) $

```python
def h(t, Ts):
    conds = [np.abs(t) < 2e-5, np.abs(t) > 1e-5]
    funds = [
        lambda x: 1,
        lambda x: np.sin(np.pi * x / Ts) / np.pi / (x / Ts)
    ]
    return np.piecewise(t, conds, funds)
```

```python
def ftn(t):
    return np.cos( 2*np.pi * t + np.pi / 4)
```

```python
t = np.linspace(-5, 15, 1000)
plt.figure(figsize=(18,9))
plt.plot(t, ftn(t), label = 'ftn(t)')

plt.xlim( np.min(t), np.max(t))
plt.legend(loc=3)

Ts = 0.3
xs = np.array( list(range(-100, 100))) * Ts
ys = ftn(xs)
plt.figure(figsize=(18,9))
plt.stem(xs, ys, 'ro-', label = 'sampled')

v = np.zeros_like(t)
for x, y in zip(xs, ys):
    v = v + y * h(t-x, Ts)
    
plt.plot(t, v, label = "reconstructed")

plt.xlim( np.min(t), np.max(t))
plt.legend(loc=3)
```
