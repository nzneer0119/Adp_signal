### Example
A train of triangular, discrete-time pulses $t[n]$ of period $ N = 11 $ has a  period <br> 
$$ \tau [n] = \begin{cases}n & 0 \leq n \leq 5\\-n + 10 & 0 \leq n \leq 5\\0 & otherwise \end{cases}$$<br>
Find the an expression for its finite difference $d[n] = t[n] - t[n-1] $

```python
def tau(n):
    conds = [((n >= 0) & (n <= 5)), ((n >= 6) & (n <= 10)), ((n < 0) | (n > 10))]
    funcs = [lambda n : n,
             lambda n : 10- n,
             lambda n : 0]
    return np.piecewise(n, conds, funcs)
```

```python
plt.figure(figsize = (12, 6))
n = np.array( list(range(-10, 30)))
plt.stem(n, tau(n))
```

```python
def t(n):
    return tau(n-22) +tau(n-11) + tau(n) + tau(n+11)
```

```python
plt.figure(figsize = (12, 6))
n = np.array( list(range(-10, 30)))
plt.stem(n, tau(n))

plt.figure(figsize = (12, 6))
plt.stem(n, t(n))
```

```python
plt.figure(figsize = (12, 6))
n = np.array( list(range(-10, 30)))
plt.stem(n, tau(n))

plt.figure(figsize = (12, 6))
plt.stem(n, t(n))

plt.figure(figsize = (12, 6))
plt.stem(n, t(n) - t(n-1))
```

### Example
Express as function of n the sample signal <br>
$ y[n] = 3r(t+3) - 6r(t+1) + 3r(t) - 3u(t-3) | _{t=0.15n}$

```python
def r(t):
    return t*np.heaviside(t, 1)

def u(t):
    return np.heaviside(t, 1)

def y(n):
    return 3*r(0.15*n+3) - 6 * r(0.15*n+1) + 3 * r(0.15*n) - 3 * u(0.15 *n -3)
```

```python
n = np.array( list(range(-30, 30)))
plt.figure(figsize = (12, 6))
plt.stem(n, y(n))
# Even
plt.figure(figsize = (12, 6))
plt.stem(n, (y(n) + y(-n))/2)
# Odd
plt.figure(figsize = (12, 6))
plt.stem(n, (y(n) - y(-n))/2)
```
