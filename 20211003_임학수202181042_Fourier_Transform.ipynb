### Fourier Transform

$ x(t) \leftrightarrow X(j\Omega) = X(s) $

$ x_3 (t) = A (u(t+ \frac {\tau}{2} - u(t+ \frac {\tau}{2}) \leftrightarrow X_3 (j\Omega) = A \tau \frac{\sin \left( \frac{\Omega \tau}{2} \right)}{\left( \frac{\Omega \tau}{2} \right)}$

$ A \leftarrow \frac{1}{\tau}  \;, \tau \leftarrow 0   $  이면, <br><br>
$ x_3 (t) = \delta (t) \leftrightarrow X_3 (j \Omega) = 1 \quad \quad \because \lim_{\theta \rightarrow 0} \frac{\sin \theta}{\theta} = 1$ <br><br>

$\delta(t)$는 모든 주파수 대를 골고루 가지고 있는 독특한 신호이다.

$ x(t) = u(t) - u(t-1) \leftrightarrow X(j \Omega) = \int_{0}^{1} {e}^{-j \Omega t}dt = \left[ \frac{1}{j \Omega} {e}^{-j \Omega t} \right] _0^1$<br><br>
$ \quad \quad \quad \quad \quad  = \frac{1}{j \Omega} \left( {e}^{-j \Omega} -1 \right) = \frac{1}{j \Omega} \left( 1- {e}^{-j \Omega} \right) =  \frac{1}{j \Omega} {e}^{- \frac{j \Omega}{2}}\left(e^{\frac{j \Omega}{2}}-e^{-\frac{j \Omega}{2}} \right)$<br><br>
$ \quad \quad \quad \quad \quad = \frac{1}{-j \Omega} e^{-\frac{j \Omega}{2}}2j \sin \left( \frac{j \Omega}{2} \right) 
= e ^{\frac{-j \Omega}{2}} \frac{2}{\Omega} \sin \left( \frac{\Omega}{2} \right) =  \frac{\sin \left( \frac{\Omega}{2} \right)e^{-\frac{j\Omega}{2}}}{\frac{\Omega}{2}} $

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def h(t):
    return np.heaviside(t, 0) - np.heaviside(t-1, 0)
def H(w):
    return np.sin( w/2) / ( w/2)
t = np.linspace(-5, 5, 1000)
plt.figure(figsize = (12,6))
plt.plot(t, h(t), label = 'h(t)')
plt.plot(t, h(2*t), label = 'h(2*t)')
plt.legend()

w = np.linspace(-50, 50, 1000)
plt.figure( figsize = (12,6))
plt.plot(w, np.abs(H(w)))
plt.plot(w, np.abs(H(w/2)/2))
```

$ X(j\Omega) = \frac{2a}{a^2 + \Omega^2}$

```python
def h(t, a):
    return np.exp(-a * np.abs(t))
def H(w, a):
    return 2*a / (a**2 + w**2)
t = np.linspace(-5, 5, 1000)
plt.figure(figsize = (12,6))
plt.plot(t, h(t, 1), label = 'a=1')
plt.plot(t, h(t, 2), label = 'a=2')
plt.legend()

w = np.linspace(-50, 50, 1000)
plt.figure( figsize = (12,6))
plt.plot(w, np.abs(H(w, 1)))
plt.plot(w, np.abs(H(w, 2)))
```

### Duality
$ x(t) \leftrightarrow X(\Omega) \quad \quad \quad \quad \left(  X(\Omega) = \int_{- \infty}^{\infty} x(t) {e}^{-j \Omega t} dt \right) $ <br><br>
$ X(t) \leftrightarrow 2 \pi x(- \Omega) $
 

$ X(\Omega) = \int_{- \infty}^{\infty} x(t) {e}^{-j \Omega t} dt $ <br><br>
$ x(t) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} X(\Omega) {e}^{j \Omega t} d\Omega = \frac{1}{2 \pi} \int_{-\infty}^{\infty} X(t^*) {e}^{j t^* t} dt^*  $ <br><br>
$ x(-\Omega) = \frac{1}{2 \pi} \int_{-\infty}^{\infty} X(t^*) {e}^{-j t^* t} dt^*  $ <br><br>
$ 2\pi x(-\Omega) = \int_{-\infty}^{\infty} X(t) {e}^{-j \Omega t} dt  $ <br><br>

$$ x(t) \leftrightarrow X(\Omega) $$<br>
$$ X(t) \leftrightarrow 2\pi x(-\Omega) $$<br>
$$ \delta(t) \leftrightarrow 1 $$<br>
$$ 1 \leftrightarrow 2 \pi \delta(\Omega) $$<br>
$$ A \leftrightarrow 2 \pi A \delta(\Omega) $$<br>

### Example
$ x_1 (t) = e^{- \mid t \mid}, \quad \quad \quad \quad -\infty < t < \infty $ <br>
$ x_2 (t) = 0.2 [r(t+5) - 2r(t) +r(t+5)] $

```python
def h(t, a):
    return np.exp(-a * np.abs(t))
def H(w, a):
    return 2*a / (a**2 + w**2)
t = np.linspace(-10, 10, 1000)
plt.figure(figsize = (12,6))
plt.plot(t, h(t, 1), label = 'a=1')
plt.plot(t, h(t, 1) * np.cos(20*t), label = 'a=2')
plt.title('Time domain')
plt.legend()

w = np.linspace(-50, 50, 1000)
plt.figure( figsize = (12,6))
plt.plot(w, np.abs(H(w, 1)))
plt.plot(w, np.abs(H(w+20, 1) + H(w-20, 1))/2)
plt.title('Frequency domain')
```

```python
def r(t):
    return t * np.heaviside(t, 0)
def h(t):
    return 0.2 * (r(t+5) - 2 * r(t) + r(t-5))
def H(w):
    return np.abs(np.sin(w * 5 /2) / (w * 5 /2)) **2
t = np.linspace(-10, 10, 1000)
plt.figure(figsize = (12,6))
plt.plot(t, h(t), label = 'a=1')
plt.plot(t, h(t) * np.cos(10*t), label = 'a=1')
plt.title('Time domain')
plt.legend()

w = np.linspace(-50, 50, 1000)
plt.figure( figsize = (12,6))
plt.plot(w, np.abs(H(w)))
plt.plot(w, np.abs(H(w+20) + H(w-20))/2)
plt.title('Frequency domain')
```
