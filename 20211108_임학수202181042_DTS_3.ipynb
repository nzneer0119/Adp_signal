## Discrete Time Signals
### Discrete -time Periodic Sinusoids

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import Image
```

```python
 Image("D:/py_WS/img/20211101_1.png")
```

```python
n = np.array(list(range(-5, 15)))
def x(n):
    return n * (np.heaviside(n, 0) - np.heaviside(n-10, 0) )
```

```python
plt.figure(figsize=(12, 6))
plt.stem(n, x(n))

plt.figure(figsize=(12, 6))
plt.stem(n, x(n-4))

plt.figure(figsize=(12, 6))
plt.stem(n, x(-n))
```

```python
plt.figure(figsize=(12, 6))
plt.stem(n, x(n))
# Even
plt.figure(figsize=(12, 6))
plt.stem(n, (x(n)+x(-n))/2)
# Odd
plt.figure(figsize=(12, 6))
plt.stem(n, (x(n)-x(-n))/2)
```

```python
plt.figure(figsize=(12, 6))
plt.stem(n, x(n))
# Even
plt.figure(figsize=(12, 6))
plt.stem(n, (x(n+3)+x(n-3)))
```

### Find even and odd component
$$ x[n] = \begin{cases}4 - n & 0 \leq n \leq 4 \\ 0 & otherwise\end{cases} $$

```python
n = np.array(list(range(-5, 15)))
def x(n):
    return (4-n) * (np.heaviside(n,1) - np.heaviside(n-4, 1))
```

```python
plt.figure(figsize=(12, 6))
plt.stem(n, x(n))
# Even
plt.figure(figsize=(12, 6))
plt.stem(n, (x(n)+x(-n))/2)
# Odd
plt.figure(figsize=(12, 6))
plt.stem(n, (x(n)-x(-n))/2)
```
