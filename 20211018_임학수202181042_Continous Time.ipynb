## Continous Time

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy import signal
import numpy as np
%matplotlib inline
from IPython.display import Image
```

#### Capacitor에 걸리는 전압 $ {v}_{c}(t)$

```python
def x(t):
    return np.cos(0.1*t) + np.cos(10*t)
```

```python
def model (y, t, x):
    return [y[1], -y[1] -y[0] + x(t)]
```

```python
t = np.linspace(0, 25, 1000)
y0 = [0, 0]
y = odeint( model, y0, t, args = (x,))
```

```python
plt.figure(figsize=(18,9))
plt.plot(t, x(t), label = 'input')
plt.legend()

plt.figure(figsize=(18,9))
plt.plot(t, y[:,0], label = 'ode')
plt.legend()
```

```python
def x(t):
    return np.cos(0.1*t) + np.cos(10*t)
def H(s):
    return 1/(s**2+s+1)
```

```python
def model (y, t, x):
    return [y[1], -y[1] -y[0] + x(t)]
```

```python
t = np.linspace(0, 25, 1000)
y0 = [0, 0]
y = odeint( model, y0, t, args = (x,))
```

```python
plt.figure(figsize=(18,9))
plt.plot(t, x(t), label = 'input')
plt.legend()

plt.figure(figsize=(18,9))
plt.plot(t, y[:,0], label = 'ode')
w =0.1
y1 = H(1j*w) * np.exp(1j *w *t)
plt.plot(t, np.real(y1), label = 'transfer _ w = 0.1')
w =10
y2 = H(1j*w) * np.exp(1j *w *t)
plt.plot(t, np.real(y2), label = 'transfer _ w = 10')
w =10
plt.plot(t, np.real(y1 + y2), label = 'transfer')
plt.legend()
```

```python
sys = signal.TransferFunction([1], [1, 1, 1])
w, mag, phase = signal.bode( sys)

plt.figure(figsize=(6,6))
plt.plot(np.real(sys.poles), np.imag(sys.poles), 'x')
plt.plot(np.real(sys.zeros), np.imag(sys.zeros), 'o')
t = np.linspace(-5, 5, 100)
plt.plot(0*t, t)

plt.xlim([-5, 5])
plt.ylim([-5, 5])

plt.figure(figsize=(18,9))
plt.semilogx(w, mag, label ='magnitude')
plt.legend()
```
