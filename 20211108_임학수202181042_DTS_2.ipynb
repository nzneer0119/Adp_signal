## Discrete Time Signals
### Discrete -time Periodic Sinusoids

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import Image
```

Periodic discrete-time sinusoids, of peiod Nm are of the form<br>
$ x[n] = A \cos \left( \frac{2 \pi m}{N} n + \theta \right) \quad \quad -\infty < n <\infty $ <br>
where the discrete frequency is $ {\omega}_{0} = \frac{2 \pi m}{N} (rad)$, for positive integers $m$ and $N$ <br>
which are not divisible by each other, and $ \theta$ is the phase angle.

```python
def x(w, n):
    return np.cos( w * n )
```

```python
m = 4
N = 10

plt.figure(figsize=(18, 9))
n = np.array(list(range(0, 50)))
plt.stem(n, x(2*np.pi*m/N, n), 'ro-')
```

```python
m = 4
N = 10
t = np.linspace(0, 50, 1000)

plt.figure(figsize=(18, 9))
n = np.array(list(range(0, 50)))
plt.stem(n, x(2*np.pi*m/N, n), 'ro-')
plt.plot(t, x(2*np.pi*m/N, t))
```

Does $x[n]$ have finite enegy, finite-power as compared with $x(t)$ when ${\Omega}_{0} = \pi$ and when ${\Omega}_{0} = 3.2 rad/sec$ (an upper approximation of $\pi$)?
