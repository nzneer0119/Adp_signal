## * Representation using Basic Signals
### * Complex Exponentials

#### * A complex exponential is a signal of the form
> $ x(t) = A e^{at} = \mid A \mid \left[ \cos(\Omega_{0}t + \theta) + j \sin(\Omega_{0}t + \theta)   \right] (-\infty < t < \infty)$
#### where $ A = \mid A \mid e^{j \theta}$ , and $ a = r + j \Omega _{0} $ are complex numbers.

### * Laplace Transform
#### * Laplace Transform
> $  L\{f\}(s) = \int_{0-}^{\infty} f(t) e^{-st} dt $

#### * Inverse Laplace Transform
> $f(t) = L^{-1}\{F\}(t) = \frac{1}{2 \pi j}\lim_{T \rightarrow \infty} \int_{\gamma - j T}^{\gamma + j T} F(s) e^{st}ds = \frac{1}{2 \pi j} \int_{\gamma - j \infty}^{\gamma + j \infty} F(s) e^{st}ds$

#### Example
> $x_{1}(t) = \cos(2\pi t)\left [u(t) - u(t-1) \right]$<br>
> $x_{2}(t) = u(t) - 2u(t-1) + u(t-2)$

```python
t = np.linspace(-2, 2, 1000)
x = np.cos(2*np.pi*t) * (np.heaviside(t,0) - np.heaviside(t-1,0))
x1 = (np.heaviside(t, 0) - np.heaviside(t-1, 0))
x2 = x - x1
```

```python
plt.figure( figsize = (12,6))
plt.grid(True)
plt.plot(t, x)
plt.plot(t, x1, label = 'unit step')
plt.plot(t, x2, label = 'continuous')
plt.legend()
```

#### Use r(t) and u(t) to represent the triangular signal A(t) / derivative.
> $ A(t) = \begin{cases}t & 0 \leq t \leq 1 \\-t +2 & 1 <t \leq 2 \\0 & othewise\end{cases}$

```python
t = np.linspace(-2, 4, 1000)
def r(t):
    return t * np.heaviside(t,0)
x = t * (np.heaviside(t,0)- np.heaviside(t-1,0)) + (2-t)*(np.heaviside(t-1,0)- np.heaviside(t-2,0))
x2 = r(t) - 2*r(t-1) + r(t-2)
```

```python
plt.figure(figsize=(12,6))
plt.grid(True)
plt.plot(t, x)
plt.plot(t, x2)
```

### Generic epresentation of Signals
> $\int_{-\infty}^{\infty} f(t) \delta(t-\tau)dt = \int_{-\infty}^{\infty} f(\tau) \delta(t-\tau)dt = f(\tau) \int_{-\infty}^{\infty} \delta(t-\tau)dt =  f(\tau) $        for any $\tau$

#### By the shifting property of the impulse function $ \delta (t)$ any signal $x(t)$ can be represented by the following generic representation:
$$ x(t) = \int_{- \infty}^{\infty} = x(\tau) \delta (t- \tau) d \tau $$ <br>
$$ y(t) = \int_{- \infty}^{\infty} = x(\tau) h (t- \tau) d \tau $$
